# Creating Visualization

# Line Chart

```python
#Load the CSV file (using raw string to avoid backslash issues)
df = pd.read_csv(r"C:\Users\SAYAN DAS\OneDrive\Desktop\SkillCraft\API_SP.POP.TOTL_DS2_en_csv_v2_2590.csv", skiprows=4)

#Filter for India
india_data = df[df['Country Name'] == 'India']

#Extract population data across all years (columns are years)
years = [str(year) for year in range(1960, 2023)]  # 19602022
population = india_data.loc[:, years].values.flatten()

#Plotting
plt.figure(figsize=(12, 6))
plt.plot(years, population, marker='o', linestyle='-', color='teal')
plt.title("India's Population Growth (19602022)", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Histogram

```python
# Load the dataset
df = pd.read_csv(r"C:\Users\SAYAN DAS\OneDrive\Desktop\SkillCraft\API_SP.POP.TOTL_DS2_en_csv_v2_2590.csv", skiprows=4)

# Extract 2022 population data and drop rows with NaN
pop_2022 = df[['Country Name', '2022']].dropna()

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(pop_2022['2022'], bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram of Country Populations in 2022", fontsize=16)
plt.xlabel("Population", fontsize=12)
plt.ylabel("Number of Countries", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Age distribution bar chart

```python
import matplotlib.pyplot as plt

# Data from your slide
age_groups = ['020 Years', '2164 Years', '65+ Years']
population_millions = [512, 807, 98]

# Plotting
plt.figure(figsize=(8, 5))
bars = plt.bar(age_groups, population_millions, color=['gold', 'dodgerblue', 'hotpink'])

# Add value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10, f'{yval}M', ha='center', va='bottom', fontsize=10)

plt.title("India's Population Distribution by Age (2022)"fontsize=16)
plt.xlabel("Age Group")
plt.ylabel("Population (in millions)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
